## 2.1.1 Sinusoid Wave
Sinusoid along the x-direction with amplitude between 0 and 1, and frequency of 4 cycles per cm

```python
N = 200 
x = np.linspace(-2,2,N)
y = x

X,Y = np.meshgrid(x,y)
R = np.sin(2*np.pi*4*X) + 0.5
A = np.zeros(np.shape(R))
A[np.where(R<0.5)] = 1.0

#remove tick marks
plt.tick_params(axis='both',bottom=False,top=False,right=False,\
                    left=False,labelbottom=False,labelleft=False)
#display as an image
plt.imshow(A, cmap = "gray")
#save actual plot
plt.savefig("sinusoid2D.png",dpi=300,bbox_inches="tight")

#save images
img = (A*255).astype(np.uint8)
for i in [".jpg",".bmp",".png",".tiff"]:
    plt.imsave("sinusoid"+i,img)

#display as a 3D surface in Cartesian coordinates
fig = plt.figure()
ax = Axes3D(fig)
ax.plot_surface(X,Y,A)
#save actual plot
plt.savefig("sinusoid3D.png",dpi=300,bbox_inches="tight")
```

## 2.1.4 Annulus
The outer circle radius should be 2cm with thickness of 0.25cm

```python
N = 200 
x = np.linspace(-2,2,N)
y = x

X,Y = np.meshgrid(x,y)
R = np.sqrt(X**2 + Y**2)
A = np.zeros(np.shape(R))
#outer circle radius is 2cm
A[np.where(R<2)] = 1.0
#thickness of outer circle is 0.25cm
A[np.where(R<1.75)] = 0.0

#remove tick marks from plot
plt.tick_params(axis='both',bottom=False,top=False,right=False,\
                    left=False,labelbottom=False,labelleft=False)
#display as an image
plt.imshow(A, cmap="gray")

#save images
img = (A*255).astype(np.uint8)
for i in [".jpg",".bmp",".png",".tiff"]:
    plt.imsave("annulus"+i,img)
```

## 2.1.5 Gaussian
Circular aperture with zero-centered Gaussian profile (standard deviation of 1cm). Circular aperture should have a radius of 1.75cm.

```python
N = 200 
x = np.linspace(-2,2,N)
y = x
r = 1.75 #radius of circle in cm

#we define a function that returns the gaussian function
def gaussian(X, sigma):
    exp = np.exp(-(X**2)/(2*sigma**2))
    return exp/(sigma*np.sqrt(2*np.pi))

X,Y = np.meshgrid(x,y)
R = np.sqrt(X**2 + Y**2)
A = np.zeros(np.shape(R))

#calculate gaussian function of R for circle within A
for i in range(X.shape[0]):
    for j in range(Y.shape[0]):
        if R[i,j] <= r:
            A[i,j] = gaussian(R[i,j],1) - 0.1
            
#remove tick marks
plt.tick_params(axis='both',bottom=False,top=False,right=False,\
                    left=False,labelbottom=False,labelleft=False)
#display as an image
plt.imshow(A, cmap="plasma")
#save actual plot
plt.savefig("gaussian2D.png",dpi=300,bbox_inches="tight")

#save images
img = (A*255).astype(np.uint8)
for i in [".jpg",".bmp",".png",".tiff"]:
    plt.imsave("gaussian"+i,img)

#display as a 3D surface in Cartesian coordinates
fig = plt.figure()
ax = Axes3D(fig)
ax.plot_surface(X,Y,A,cmap="plasma")
#save actual plot
plt.savefig("gaussian3D.png",dpi=300,bbox_inches="tight")
```

## 2.2 Olympics Logo
Simulate the Olympics logo with overlapping colored circles

```python
N = 500
x = np.linspace(-15,15,N)
y = x

X,Y = np.meshgrid(x,y)
#to have a white background, we fill the 2D-array with 255 values
Rd, Gn, Bl = np.full((N,N),255),np.full((N,N),255),np.full((N,N),255)

#draw colored circles
Rt, Rc, deg, r = 3, 4, 30, 3.5
xt, yt = Rt*np.cos(deg*np.pi/180), Rt*np.sin(deg*np.pi/180)

#to achieve the desired color, we need to subtract the appropriate
#RGB layers (this is similar to setting a layer to 0)

Black = np.sqrt((X)**2 + (Y)**2)
#color decimal
Rd[np.where(Black<Rc)]=0
Gn[np.where(Black<Rc)]=0
Bl[np.where(Black<Rc)]=0
#inner radius
Rd[np.where(Black<r)]=255
Gn[np.where(Black<r)]=255
Bl[np.where(Black<r)]=255

Yellow = np.sqrt((X+1.75*xt)**2 + (Y-3*yt)**2)
#color decimal
Bl[np.where(Yellow<Rc)]=0
#inner radius
Bl[np.where(Yellow<r)]=255

Green = np.sqrt((X-1.75*xt)**2 + (Y-3*yt)**2)
#color decimal
Rd[np.where(Green<Rc)]=0
Bl[np.where(Green<Rc)]=0
#inner radius
Rd[np.where(Green<r)]=255
Bl[np.where(Green<r)]=255

Blue = np.sqrt((X+3.5*xt)**2 + (Y)**2)
#color decimal
Gn[np.where(Blue<Rc)]=0
Rd[np.where(Blue<Rc)]=0
#inner radius
Gn[np.where(Blue<r)]=255
Rd[np.where(Blue<r)]=255

Red = np.sqrt((X-3.5*xt)**2 + (Y)**2)
#color decimal
Gn[np.where(Red<Rc)]=0
Bl[np.where(Red<Rc)]=0
#inner radius
Gn[np.where(Red<r)]=255
Bl[np.where(Red<r)]=255

I = np.zeros((N,N,3))
I[...,0] = Rd
I[...,1] = Gn
I[...,2] = Bl

fig = plt.figure()
#remove tick marks
plt.tick_params(axis='both',bottom=False,top=False,right=False,\
                    left=False,labelbottom=False,labelleft=False)
#display image
plt.imshow(I.astype(np.uint8))
#save actual plot
plt.savefig("olympics2D.png",dpi=300,bbox_inches="tight")

#save images
img = (I*255).astype(np.uint8)
for i in [".jpg",".bmp",".png",".tiff"]:
    plt.imsave("olympics"+i,img)
```
